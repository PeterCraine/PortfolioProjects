```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
URL = 'https://www.amazon.com/dp/B0CFQPQSPJ?ref_=cm_sw_r_cso_cp_apin_dp_G1S9C9NMCVEXMB07KCRK&cafHiResImg=1&language=en-US&th=1'
options = Options()

try:
    print("Starting WebDriver...")
    driver = webdriver.Chrome(options=options)
    print("Navigating to URL...")
    driver.get(URL)
    
    print("Extracting Total Ratings...")
    total_ratings_element = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.ID, "acrCustomerReviewText"))
    )
    total_ratings = total_ratings_element.text.strip()
    print("Total Ratings:", total_ratings)
    print("Extracting Product Title...") 
    title_element = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.ID, "productTitle"))
    )
    title = title_element.text.strip()
    print("Title:", title)

    print("Extracting Product Price...")
    price_whole = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.CLASS_NAME, "a-price-whole"))
    ).text
    price_fraction = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.CLASS_NAME, "a-price-fraction"))
    ).text
    price = price_whole + '.' + price_fraction
    print("Price:", price)

    print("Extracting Average Rating...")
    avg_rating = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.XPATH, "//span[@data-hook='rating-out-of-text']"))
    ).text
    print("Average Rating:", avg_rating)

    print("Extracting Star Percentages...")
    ratings_percent = WebDriverWait(driver, 10).until(
        EC.presence_of_all_elements_located((By.XPATH, "//tr[@class='a-histogram-row a-align-center']//a[contains(@aria-label, 'percent of reviews')]"))
    )
    
    # Store percentages in a dictionary
    star_percentages = {}
    for i, rating in enumerate(ratings_percent, start=1):
        percentage = rating.text.replace('%', '').strip()
        star_percentages[5-i+1] = percentage
    
    # Now you can print the stored percentages whenever you want
    print("\nPrinting Star Percentages:")
    for star, percentage in star_percentages.items():
        print(f"{star} Star: {percentage}%")

except Exception as e:
    print("An error occurred:", e)

finally:
    print("Quitting WebDriver...")
    driver.quit()
```

```python
import csv
import time
import datetime
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By

def send_mail():
    print("send_mail")
    pass

def check_price():
    URL = 'https://www.amazon.com/dp/B0CFQPQSPJ?ref_=cm_sw_r_cso_cp_apin_dp_G1S9C9NMCVEXMB07KCRK&cafHiResImg=1&language=en-US&th=1'
    options = Options()

    try:
        driver = webdriver.Chrome(options=options)
        driver.get(URL)

        # Extracting Product Title
        title_element = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.ID, "productTitle"))
        )
        title = title_element.text.strip()

        # Extracting Prices
        price_whole = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.CLASS_NAME, "a-price-whole"))
        ).text.replace(',', '')  # Remove commas
        price_fraction = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.CLASS_NAME, "a-price-fraction"))
        ).text
        price = float(price_whole + '.' + price_fraction)  # Convert to float
        
        # Extracting Total Ratings
        total_ratings_element = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.ID, "acrCustomerReviewText"))
        )
        total_ratings = total_ratings_element.text.strip()
        
        # Extracting Average Ratings
        avg_rating = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.XPATH, "//span[@data-hook='rating-out-of-text']"))
        ).text

        # Extracting Rating Percentages
        ratings_percent = WebDriverWait(driver, 10).until(
            EC.presence_of_all_elements_located((By.XPATH, "//tr[@class='a-histogram-row a-align-center']//a[contains(@aria-label, 'percent of reviews')]"))
        )
        star_percentages = {}
        for i, rating in enumerate(ratings_percent, start=1):
            percentage = rating.text.replace('%', '').strip()
            star_percentages[5-i+1] = percentage

    except Exception as e:
        print("An error occurred:", e)

    finally:
        driver.quit()

    today = datetime.date.today()
    header = ['Title', 'Price', 'Total Ratings', 'Average Ratings', '5 Star Percent', '4 Star Percent', '3 Star Percent', '2 Star Percent', '1 Star Percent', 'Date']
    data = [title, price, total_ratings, avg_rating, star_percentages[5], star_percentages[4], star_percentages[3], star_percentages[2], star_percentages[1], today]
    file_path = 'AmazonGamingLaptop2023.csv'

    # Write data to CSV file
    try:
        with open(file_path, 'x', newline='', encoding='utf-8') as f:
            writer = csv.writer(f)
            writer.writerow(header)
    except FileExistsError:
        pass

    with open(file_path, 'a', newline='', encoding='utf-8') as f:
        writer = csv.writer(f)
        writer.writerow(data)

    if price < 1000:
        send_mail()

while True:
    check_price()
    time.sleep(86400)
```

```python
import smtplib

def send_mail():
    smtp_server = 'smtp.gmail.com'
    smtp_port = 465
    sender_email = 'Insert Email Here'  # Replace with your Gmail address
    receiver_email = 'Insert Email Here'  # Replace with receiver's email
    app_password = 'XXXXXXXX'  # Replace with your App Password
    
    subject = "The Laptop you want is below $1,000! Now is your chance to buy!"
    body = "XXXX, This is the moment we have been waiting for. Now is your chance to pick up the Gaming Laptop of your dreams. Don't mess it up! Link here: https://www.amazon.com/dp/B0CFQPQSPJ?ref_=cm_sw_r_cso_cp_apin_dp_G1S9C9NMCVEXMB07KCRK&cafHiResImg=1&language=en-US&th=1"
    msg = f"Subject: {subject}\n\n{body}"
    
    try:
        # Connect to the SMTP server using SSL
        server = smtplib.SMTP_SSL(smtp_server, smtp_port)
        server.ehlo()
        
        # Login to your email account
        server.login(sender_email, app_password)
        
        # Send the email
        server.sendmail(sender_email, receiver_email, msg)
        
        print('Email sent successfully!')

    except Exception as e:
        print('Failed to send email:', e)

    finally:
        # Always close the server connection when done
        server.quit()
```
